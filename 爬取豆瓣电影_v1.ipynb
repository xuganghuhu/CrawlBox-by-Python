```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

import requests
```

```python
from bs4 import BeautifulSoup
import urllib
```

```python
html='https://movie.douban.com/'
obj=urllib.request.urlopen(html)
soup=BeautifulSoup(obj,'html.parser')
```

```python
txt=soup.find_all('img')

for each in txt:
    print(each)
```

```python
name=[]
txt=soup.find_all('img')

for each in txt:
    print(each['alt'])
    name.append(each['alt'])
```

```python
adress=[]
txt=soup.find_all('img')

for each in txt:
    adress.append(each['src'])
```

```python
dat=pd.DataFrame(name,adress[0:47])
```

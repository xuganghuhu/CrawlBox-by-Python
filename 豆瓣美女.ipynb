```python
from bs4 import BeautifulSoup 
import urllib
import urllib.request
import pandas as pd
```

```python
x=0

url_bas='http://www.mmjpg.com'
for i in range(5):
    url=url_bas+'/home/'+str(i)
    html=urllib.request.urlopen(url)
    soup=BeautifulSoup(html,'html.parser')
    tar=soup.find_all('li')

    for each in tar:
        link=each.a['href']
        print(each.a['href'])
        
        urllib.request.urlretrieve(link,'/Users/huxugang/Github/meizi/%s.jpg'%x)
        x+=1
        print('正在下载第%s张'%x)
        
```

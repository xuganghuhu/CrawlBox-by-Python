```python
from bs4 import BeautifulSoup 
import urllib
import urllib.request
import pandas as pd
import numpy as np
```

```python
url='https://gaokao.chsi.com.cn/zsgs/bssnlqmd.do?method=listStu&year=2018&yxdm=10001'
html=urllib.request.urlopen(url)
soup=BeautifulSoup(html,'html.parser')

tar=soup.find_all('td')

obj=[]
for each in tar:
    #print(each.text)
    obj.append(each.text)
```

```python
name=[]
for i in np.arange(0,1560,8):
    name.append(obj[i])

gender=[]
for i in np.arange(0,1560,8):
    gender.append(obj[i+1])

province=[]
for i in np.arange(0,1560,8):
    province.append(obj[i+2])

school=[]
for i in np.arange(0,1560,8):
    school.append(obj[i+3])

feature=[]
for i in np.arange(0,1560,8):
    feature.append(obj[i+4])

rating=[]
for i in np.arange(0,1560,8):
    rating.append(obj[i+5])

rating2=[]
for i in np.arange(0,1560,8):
    rating2.append(obj[i+6])

major=[]
for i in np.arange(0,1560,8):
    major.append(obj[i+7])
  
```

```python
dat=pd.DataFrame({'name':name,'gender':gender,'province':province,'school':school,'feature':feature,'rating':rating,'rating2':rating2,'major':major})
dat=dat[['name','gender','province','school','feature','rating','rating2','major']]
```

```python
## 豆瓣电影top250

name=[]
star=[]
brief=[]
inf=[]
other=[]
starNum=[]

url_bas='https://movie.douban.com/top250?start='
for i in range(0,250,25):
    url=url_bas+str(i)+'&filter='

    html=urllib.request.urlopen(url)
    soup=BeautifulSoup(html,'html.parser')

    tar_name=soup.find_all('div',class_='hd')         #爬取电影名
    for each in tar_name:
        name.append(each.a.span.text)
        #print(each.a.span.text)
    
    tar_star=soup.find_all('span',class_='rating_num')      #爬取豆瓣评分
    for each in tar_star:
        star.append(each.text)
        #print(each.text)
    
    tar_brief=soup.find_all('span',class_='inq')      #爬取一句话简介
    for each in tar_brief:
        brief.append(each.text)
        #print(each.text)
    
    tar_other=soup.find_all('span',class_='other')      #爬取别名
    for each in tar_other:
        other.append(each.text)
        #print(each.text)
    
    tar_inf=soup.find_all('div',class_='bd')      #爬取详细信息
    for each in tar_inf:
        inf.append(each.p.text)
        #print(each.text)
    
for i in np.arange(0,260,26):     #详细信息里有不想要的“豆瓣”字样
    inf.pop(i)    
        
```

```python
dat=pd.DataFrame({'name':name,'star':star,'brief':brief,'other_name':other,'info':inf})
dat=dat[['name','star','other_name','info','brief']]
```

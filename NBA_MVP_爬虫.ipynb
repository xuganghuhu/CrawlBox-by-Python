```python
from bs4 import BeautifulSoup 
import urllib
import urllib.request
import pandas as pd
```

```python
## NBA爬虫
url='http://www.stat-nba.com/award/item0.html'

def crawll(url):
    html=urllib.request.urlopen(url)
    bs=BeautifulSoup(html.read(),'html.parser')
    return(bs)
    
bs_obj=crawll(url)

mygilr=bs_obj.find_all('td')

llist=[]
for girl in mygilr:
    link=girl.get_text('href')
    llist.append(link)
```

```python
#球员名字
sid=1
name=[]
while sid <=len(llist)-1:
    name.append(llist[sid])
    sid+=24
    
#场均得分
sid=23
points=[]
while sid <=len(llist)-1:
    points.append(llist[sid])
    sid+=24
    
#篮板
sid=15
rebond=[]
while sid <=len(llist)-1:
    rebond.append(llist[sid])
    sid+=24
    
#助攻
sid=18
asist=[]
while sid <=len(llist)-1:
    asist.append(llist[sid])
    sid+=24 
    
#抢断
sid=19
steal=[]
while sid <=len(llist)-1:
    steal.append(llist[sid])
    sid+=24 

```

```python
pd.DataFrame({'name':name,'points':points,'rebond':rebond,'asist':asist,'steal':steal})
```
